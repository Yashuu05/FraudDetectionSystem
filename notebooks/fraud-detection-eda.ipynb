```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fraud_count_df = pd.DataFrame(df['isFraud'].value_counts().reset_index())
fraud_count = fraud_count_df.loc[1,'count'].astype(int)
fraud_percent = (fraud_count / fraud_count_df.loc[0,'count'].astype(int)) * 100
print(f"fraud count percent = {fraud_percent:.2f} %")
```

```python
def find_frauds():
    return df[df['isFraud']==1]
```

```python
fraud_type_count = df.groupby("type").apply(find_frauds).sum().reset_index()
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
sns.barplot(data=fraud_type_count, x=fraud_type_count['type'], y=fraud_type_count['isFraud'], color='#FF9A86')
plt.title("Fraud count for each transaction type", loc='center', fontsize=14, fontweight='bold')
plt.xlabel(f"<----- type ----->")
plt.ylabel(f"<----- count ------>")
plt.tight_layout()
plt.show()
```
